```python
import modulo as md
```

```python
#TESTE DA FUNÇÃO CRIASTRING
palavra = md.criaString()
print(f'Palavra gerada: {palavra}')
```

```python
#TESTE DA FUNÇÃO CRIAPOP
pop = md.criaPop(palavra)
print(f'População inicial: {pop}')
```

```python
#TESTE DA FUNÇÃO MUTATION
pop = md.mutation(pop, md.MUTATION_RATE)
print(f'População final: {pop}')
```

```python
#TESTE DA FUNÇÃO ANALISACASO
fitness = md.analisaCaso('ME*THINKS*IT*IS*LIKE*A*WEASEL')  
print(f'"ME*THINKS*IT*IS*LIKE*A*WEASEL" fitness: {fitness}')
fitness = md.analisaCaso(pop[0])  
print(f'"{pop[0]}" fitness: {fitness}')
```

```python
#TESTE DA FUNÇÃO POPFITNESS
pop_fitness = md.popFitness(pop)
print(pop_fitness)
```
